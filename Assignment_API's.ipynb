```python
# Dependencies and Setup
#import openweathermapy.core as owm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests 
import time
from configs import api_key
import time

# Import API key


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
coord = []
cit_y = []
cities = []
# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=3500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=3500)
coord = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in coord:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cit_y:
        cit_y.append(city.title())

# Print the city count to confirm sufficient count
print(len(cit_y))
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"

units = "imperial"

#units = ["metric", "imperial"]

# Build partial query URL
#print(cities)
lat=[]
temp=[]
humi=[]
cloud=[]
wind=[]
Country=[]
Date=[]
long=[]
print(list)
print("Beginning Data Retrieval")
print("----------------------------------------")
i=0
j=0
z=2500
for city in cit_y:
    if z > 0:
        z-=1
        try:
            query_url = url + "appid=" + api_key + "&units=" + units + "&q=" + city
            response = requests.get(query_url).json()
            lat.append(response['coord'][ 'lat'])
            temp.append(response['main']['temp_max'])
            humi.append(response['main']['humidity'])
            cloud.append(response['clouds']['all'])
            wind.append(response['wind']['speed'])
            Country.append(response["sys"]["country"])
            Date.append(response["dt"])
            long.append(response['coord']['lon'])
            i+=1
            cit_y.pop(0)
            cities.append(cit_y.pop(0))
            print(f'Processing Record {i} of set {j} | for {city}')
            print(query_url)
            if i == 50:
                time.sleep(60)
                j+=1
                i=0
        except:
            continue
```

```python
weather_dict = {
    "City": cities,
    "Cloudiness": cloud,
    "Country": Country,
    "Date": Date,
    "Humidity": humi,
    "Latitute": lat,
    "Longitute": long,
    "Max Temp": temp,
    "Wind": wind
}
weather_data_dropped = pd.DataFrame(weather_dict) 
weather_data = weather_data_dropped.drop_duplicates('City')
weather_data.head(10)
weather_data.to_csv(r'../starter_code/WeatherDataFrame.csv')
#for x in range(len(cities)):
    #storage = owm.get_current(x, **settings)
    #Current_weather_city[x] = storage[x]
```
